# Python中的数据结构（二）
Python中常见的数据结构有六种，包括：Number(数值), String(字符串), List(列表), Tuple(元组), Dictionary(字典), Set(集合)

|      | 有序 | 无序 |
| ---- | ---- | ---- |
| 可变 | List | Dictionary, Set |
| 不可变 | String, Tuple | Number|

## 2.List 与 Tuple

```python
# Python3 - List(列表)：有序序列，其中的元素可以是任意类型；可以随时修改数据。
classmates = ["Michael", "Bob", "Terry"]    # 变量classmates就是一个list。
print(classmates)
print(len(classmates))                      #用len()函数可以获得list元素的个数

print(classmates[0])                        # 用索引来访问list中每一个位置的元素，记得索引是从0开始的

print(classmates[-1])                       # 如果要取最后一个元素，除了计算索引位置外，还可以用-1做索引，直接获取最后一个元素

classmates[1] = "Sarah"                     #要把某个元素替换成别的元素，可以直接赋值给对应的索引位置
print(classmates)

age = [16, 17, 18]
print(classmates + age)                     #列表+（拼接）
print(age * 3)                              #列表*（重复）
```

```python
# 列表的常用操作
classmates = ["Michael", "Bob", "Terry"]    # 变量classmates就是一个list。

classmates.append('Adam')                   # list是一个可变的有序表，所以，可以往list中追加元素到末尾
print('list.append() = ', classmates)

classmates.insert(1, "Jack")                # 也可以把元素插入到指定的位置，比如索引号为1的位置（即第2个位置）
print('list.insert() = ', classmates)

classmates.extend(['Terry', 'Sean'])        # 将可迭代对象的每一个元素逐个插入列表的尾部
print('list.extend() = ', classmates)

classmates.pop()                            # 要删除list末尾的元素，用pop()方法
print('list.pop() = ', classmates)

classmates.pop(1)                           # 要删除指定位置的元素，用pop(i)方法，其中i是索引位置
print('list.pop(1) = ', classmates)

classmates.remove('Terry')                  # 删除列表中第一个出现指定元素（参数对象）的位置
print('list.remove("Terry") = ', classmates)

print('list.index() = ', classmates.index('Bob'))              # 查找指定元素第一次出现指定元素（参数对象）的位置

age = [16, 17, 19, 15, 18]
age.sort()                                  # 对列表进行排序，默认为升序              
print('age.sort() = ', age)

age.reverse()                               # 将列表中的元素进行逆置（直接修改列表本身）
print('age.reverse() = ', age)

age.append(18)
print('age.count(18) = ', age.count(18))    # 返回指定元素出现的次数

dir(age)                                    # dir：查看对象的内置方法和属性
```

```python
# Python3 - Truple(元组)：有序序列（和列表类似），不可变（数值一旦创建不可被修改，和字符串类似）
mates = ('Michael', 'Bob', 'Tracy')     #创建一个元组mates
print("mates = ", mates)
print("mates[0] = ", mates[0])

# 元组中可以同时包含不同类型的元素
sth = (1,True,0,'Apple',[1,2,[1,3]])
print('输出一个多类型元素的元组 = ',sth)

# 当你定义一个tuple时，在定义的时候，tuple的元素就必须被确定下来
# 定义一个空的tuple，可以写成()
t = ()
print('输出一个空元组 = ',t)

# 可变的tuple：tuple不能变，它没有append()，insert()这样的方法，但可以改变其内部的元素内容
t = ("a", "b", ["c", "d"])
t[2][0] = "X"
t[2][1] = "Y"
print('输出("a", "b", ["c", "d"])修改其中的元素后的元组 = ',t)
t[2].append('Z')
print("输出新的t = ", t)

# 通过索引访问
L = [['Apple', 'Google', 'Microsoft'], ['Java', 'Python', 'Ruby', 'PHP'], ['Adam', 'Bart', 'Lisa']]
print("L[0][0] = ", L[0][0])
print("L[1][1] = ", L[1][1])
print("L[2][2] = ", L[2][2])
print("L[-1][-1] = ",L[-1][-1])
print("L[-2][0:3] = ",L[-2][0:3])       # 通过切片访问
print("L[-2][:3] = ",L[-2][:3])         # 通过切片访问
print("L[0][:-1] = ",L[0][:-1])         # 通过切片访问

# 通过 + 添加元素
a=(1,2,3)
print('id a = ',hex(id(a)))                         # 查看变量a的内存地址
a+=(4,5,6)
print('id a = ',hex(id(a)))                         # 再次查看变量a的内存地址，此时的a与之前的a的地址不同了，说明已经不是一个a了，元组是不可改变的数据类型 
print("a+=(4,5,6) = ", a)
# 通过 * 添加元素
b=a*3
print("a*3 = ", b)
```

```python
# 元组的常用操作
tpl = (1,3,5,3,7,6,8,2,0,1)
print("1 in tpl = ", 1 in tpl)                  # 判断一个元素是否存在于元组中
print("len(tpl) = ", len(tpl))                  # 获取元组的长度
print("max(tpl) = ", max(tpl))                  # 获取元组中最大元素
print("min(tpl) = ", min(tpl))                  # 获取元组中最小元素
print("sum(tpl) = ", sum(tpl))                  # 计算元组中元素的和
print("tpl.index(3) = ", tpl.index(3))          # 获取元组中元素3的下标
print("tpl.count(3) = ", tpl.count(3))          # 获取元组中元素3的个数

# 删除元组，只能使用del清空元组
a=(1,2,3,4)
del a
# 元组的一些坑
# (1) tuple陷阱
a = [1]                                 # 定义一个元素的列表
print("type([1]) = ", type(a))
b = (1)                                 # 定义一个元素的元组？ - 这是错误的。
print("type((1)) = ", type(b))          # 括号()既可以表示tuple，又可以表示数学公式中的小括号，这就产生了歧义，因此这里的(1)在python中是数学运算符号，因此输出的是int类型
print(type((1,)))                       # 定义一个元素的元组，其正确的做法是在元素后面添加一个逗号

# dir：查看对象的内置方法和属性
dir(tpl)
```
